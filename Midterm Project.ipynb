Counting the frequencies of the numbers using the dictionary in Python

    #Find the frequency of each distinct element using a dictionary.
    #Plot the frequency results, e.g. using histogram 
    #Store this dictionary results into json file and store it in your local computer. 

```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import json
import pandas as pd
```

```python
file = open("numbers.txt", "r") #read file
contents = file.read()
print(contents)
```

```python
new_contents = contents.replace("\n", ",") #replace line breaks with commas
print(new_contents)
```

```python
list_numbers = new_contents.split(',') #create list of iterables 
print(list_numbers)
```

```python
number_count = {} #create empty dictionary
number_count
```

```python
for number in list_numbers: #count frequencies using a dictionary
    number = int(number) #turn strings into integers
    if number in number_count:
        number_count[number] += 1
    else:
        number_count[number] = 1
print(number_count)
```

```python
number_count2 = {} #create new sorted dictionary
for i in sorted(number_count):
    number_count2[i] = number_count[i]
print(number_count2)
```

```python
for key, value in sorted(number_count2.items()): #print frequencies
    print(f'{key} is in the file {value} times') 
```

```python
#Plot the frequency results, e.g. using histogram
```

```python
number_count2.values()
```

```python
freq = number_count2.values()
plt.hist(freq, bins = [5, 10, 15, 20, 25, 30])
plt.xlabel("Freqency")
plt.ylabel("Count")
plt.title("No. of Times an Integer Appears in the File")
plt.show()
```

```python
#Store this dictionary results into json file and store it in your local computer.
```

```python
with open("./numfreqdict.json", "w") as file2: #stored in same file as notebook
    json.dump(number_count2, file2)

f = open("numfreqdict.json", "r")
f.read()
```

```python
df = pd.read_csv('All-Amazon-Orders.csv') #read file
df.head()
```

```python
df.shape #shape of DataFrame
```

```python
df = df.fillna(value = 0) #fill missing values
df.head(2)
```

```python
new_df = df.columns #new variable that accesses all columns in DataFrame
new_df
```

```python
df[new_df] = df[new_df].replace({'\$' : ''}, regex = True) #replace all $ in DataFrame
df.head(2)
```

```python
df['Total Charged'] = df['Total Charged'].astype(float) #convert object to float for analysis
print(df.dtypes)
```

```python
a = df['Total Charged'].sum()
b = df['Total Charged'].mean()
c = df['Total Charged'].median()
d = df['Total Charged'].max()
e = df['Total Charged'].min()
f = df['Total Charged'].std()
print(f'Total sum is {a}\nTotal mean is {b}\nTotal median is {c}\nTotal max is {d}\nTotal min is {e}\nTotal std is {f}')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date']) #change column into dates
df.head(2)
```

```python
df.groupby('Order Date').sum()['Total Charged']
```

```python
df['year'] = pd.DatetimeIndex(df['Order Date']).year
df['month'] = pd.DatetimeIndex(df['Order Date']).month
df.head(2)
```

```python
full_summary = df.groupby(['year', 'month']).agg(['sum', 'min', 'median', 'max', 'mean', 'std'])['Total Charged']
pd.set_option('max_rows', None)
print(full_summary)
```

```python
monthly_avg = df.groupby(['year', 'month']).mean()['Total Charged']
monthly_avg
```

```python
yearly_sum = df.groupby('year').sum()['Total Charged']
yearly_sum
```

```python
#pie chart
mylabels = ["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]
myexplode = [0.2, 0.4, 0.1, 0.1, 0.1, 0.1, 0.2, 0.3, 0.1, 0.1, 0.1, 0.1]
plt.rcParams["figure.figsize"] = (10, 7)
plt.pie(yearly_sum, labels = mylabels, explode=myexplode)
plt.legend(title="2011-2022 Total Amount Charged Divided by Year", bbox_to_anchor=(1.2, 1.0), loc='upper left')
plt.show() 
```

```python
#multiple line comparison
pivoted = pd.pivot_table(df, values='Total Charged', columns='year', index='month')
pivoted.plot()
plt.title('Total Amount Charged per Month by Year')
plt.ylabel('Amount Spent in $')
plt.xlabel('Month')
plt.grid()
```
