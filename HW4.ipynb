```python
# Question 1.1 Iterating through a Multidimensional array’s Elements. Using nested for loop, to print out the array in Q1
for row in array1:
    for column in row:
        print(column, end=' ')
    print()
```

```python
# Questions 5: Read each student’s grade from the marks dictionary. Then plot a bar chart and a pie chart for each 
#student’s grade. Label student’s name at each plot. Add plot title. Add legend for the pie chart. 
marks = {'Andy': 88, 'Amy': 66, 'James': 90, 'Jules': 55, 'Arthur': 77}

for name, grade in marks.items():
    print(f'{name}: {grade}')
    
names = list(marks.keys())
grades = list(marks.values())
colors = ["r", "orange", "yellow", "g", "b"]

plt.subplot(1, 2, 1)
plt.bar(names, grades, color = "purple")

plt.subplot(1, 2, 2)
plt.pie(grades, labels = names, colors = colors)
plt.legend(title = "Grades", bbox_to_anchor = (1.0, 1.0), loc = 'upper left')

plt.suptitle("Gradebook")
plt.show()
```

```python
# QUestion 6: Practice subplots() function, draw a subplot with 2 rows and 3 columns. Add each plot title.

fig, ax = plt.subplots(2, 2, figsize=(15,12))
# One subplot has multiple lines
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([5, 6, 7, 8, 9])

plt.subplot(2, 3, 1)
plt.plot(x1, y1, x2, y2)
plt.title("Multiple Lines")

# one subplot is a bar chart
x = np.array(["Jan.", "Feb.", "Mar.", "Apr.", "May", "June"])
y = np.array([1, 7, 5, 1, 6, 5])

plt.subplot(2, 3, 2)
plt.bar(x, y)
plt.xlabel("Month")
plt.ylabel("No. of Books Read")
plt.title("Bar Chart")

# one subplot is a pie chat
y = np.array([3, 1, 4])
mylabels = np.array(["Amy", "Jill", "George"])

plt.subplot(2, 3, 3)
plt.pie(y, labels = mylabels)
plt.legend(title = "Pieces of Pizza Eaten", bbox_to_anchor = (1.05, 1.0), loc = 'upper left')
plt.title("Pie Chart")

# one subplot has a grid, you define it yourself
x = np.array([1, 2, 3])
y = np.array([90, 87, 94])

plt.subplot(2, 3, 4)
plt.plot(x, y)
plt.grid()
plt.xlabel("Day")
plt.ylabel("Temp.")
plt.title("Plot with Grid")

# one subplot is a scatter plot
x = np.array([56, 16, 28, 32, 37, 21, 45])
y = np.array([2, 1, 3, 2, 4, 2, 1])

plt.subplot(2, 3, 5)
plt.scatter(x, y)
plt.xlabel("Age")
plt.ylabel("No. of Pets Owned")
plt.title("Scatter Plot")

# one subplot is a histogram plot
x = [1, 5, 6, 5, 4, 5, 2, 3, 5, 2, 2, 4, 1]

plt.subplot(2, 3, 6)
plt.hist(x)
plt.ylabel("Papercuts Acquired by Librarians (in a week)")
plt.title("Histogram")

plt.subplots_adjust(wspace = 0.7, hspace = 0.7)
plt.show()
```
